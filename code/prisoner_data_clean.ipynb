```python
#import modules
import pandas as pd
```

```python
#Path the file locations
prisoners = "../data/crime/usa_crime_rate.csv"
```

```python
#Convert csv files into dataframes
prisoners_df = pd.read_csv(prisoners, index_col = 0)
prisoners_df = prisoners_df.T
prisoners_df
```

```python
#Get the states into a list for future referance
state_list = [ "Alabama", "Arkansas", 
            "Arizona", "California", "Colorado", "Connecticut", 
            "District of Columbia", "Delaware", "Florida", "Georgia", 
            "Hawaii", "Iowa", "Idaho", "Illinois", "Indiana", "Kansas", 
            "Kentucky", "Louisiana", "Massachusetts", "Maryland", "Maine", "Michigan", 
            "Minnesota", "Missouri", "Mississippi", "Montana", "North Carolina", 
            "North Dakota", "Nebraska", "New Hampshire", "New Jersey", "New Mexico", 
            "Nevada", "New York", "Ohio", "Oklahoma", "Oregon", "Pennsylvania", 
            "Rhode Island", "South Carolina", "South Dakota", "Tennessee", 
            "Texas", "Utah", "Virginia", "Vermont", "Washington", 
            "Wisconsin", "West Virginia", "Wyoming"]
```

```python
#Locate the first state in the data
prisoners_data = pd.DataFrame(prisoners_df.T.loc["Alaska",:])

#container for loop appending
state_cont = []

#loop to get repeating state name in column and setting up new loop
for x in range(len(prisoners_data)):
        state_cont.append(prisoners_data.columns[0])
prisoners_data["State"] = state_cont
prisoners_data = prisoners_data.set_index(["State",prisoners_data.index]) 
prisoners_data.columns=["Prisoners Rate of change"]

#Iterate through all states to get data in a preferred format for mongodb
for state in state_list:

    prisoners_new = pd.DataFrame(prisoners_df.T.loc[state,:])

    state_cont = []

    for x in range(len(prisoners_new)):
        state_cont.append(prisoners_new.columns[0])
    prisoners_new["State"] = state_cont
    prisoners_new = prisoners_new.set_index(["State",prisoners_new.index])
    prisoners_new.columns=["Prisoners Rate of change"]
    prisoners_data = prisoners_data.append(prisoners_new)
prisoners_data
```

```python
prisoners_data.index.names = ["State" , "Date"]
prisoners_data
```

```python
#Save finished file into a csv
prisoners_data.to_csv("../data/prisoners_rate.csv")
```
