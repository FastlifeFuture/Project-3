```python
#import dependancies
import pandas as pd
```

```python
#Path to file
happy = "../data/happiness.csv"
```

```python
#Convert csv data into a pandas dataframe
happy_df = pd.read_csv(happy)
happy_df
```

```python
happy_df["GDP Growth Rate"] = happy_df["GDP"].pct_change()*100
happy_df = happy_df.fillna(0)
happy_df
```

```python
happy_df["State"][0]
```

```python
i = 0
total = happy_df["State"][i]
gdp = []

# demonstrating a basic while-loop
if total == total:
    happy_df["GDP Growth Rate"] = happy_df["GDP"].pct_change()*100
    i += 1
    gdp.append(happy_df)
happy_df


```
