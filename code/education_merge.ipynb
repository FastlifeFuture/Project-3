```python
#import modules
import pandas as pd
```

```python
#Path the file locations of the csv files
college = "../data/original/College_Education_States_Annual.csv"
high_school = "../data/original/High_School_Education_States_Annual.csv"
```

```python
#Convert csv files into dataframes
college_df = pd.read_csv(college, index_col =0)
hs_df = pd.read_csv(high_school, index_col =0)

#Convert data into a more usable way for merging
t_college_df = college_df.stack(level=0).unstack(level=0)
t_hs_df = hs_df.stack(level=0).unstack(level=0)

#Reset index for proper formatting 
college_reset = t_college_df.reset_index()
hs_reset = t_hs_df.reset_index()
```

```python
#Get the states into a list for future referance
state_list = [ "Alabama", "Arkansas", 
            "Arizona", "California", "Colorado", "Connecticut", 
            "District of Columbia", "Delaware", "Florida", "Georgia", 
            "Hawaii", "Iowa", "Idaho", "Illinois", "Indiana", "Kansas", 
            "Kentucky", "Louisiana", "Massachusetts", "Maryland", "Maine", "Michigan", 
            "Minnesota", "Missouri", "Mississippi", "Montana", "North Carolina", 
            "North Dakota", "Nebraska", "New Hampshire", "New Jersey", "New Mexico", 
            "Nevada", "New York", "Ohio", "Oklahoma", "Oregon", "Pennsylvania", 
            "Rhode Island", "South Carolina", "South Dakota", "Tennessee", 
            "Texas", "Utah", "Virginia", "Vermont", "Washington", 
            "Wisconsin", "West Virginia", "Wyoming"]
```

```python
#Locate the first state in our data
college_data = pd.DataFrame(college_df.T.loc["Alaska",:])

#container for loop appending
state_cont = []

#Loop to get repeating state name in column and setting up new loop
for x in range(len(college_data)):
    state_cont.append(college_data.columns[0])
college_data["State"] = state_cont
college_data = college_data.set_index(["State",college_data.index]) 
college_data.columns=["College"]

#Iterate through all states to get data in a preferred format for mongodb
for state in state_list:

    college_new = pd.DataFrame(college_df.T.loc[state,:])

    state_cont = []

    for x in range(len(college_new)):
        state_cont.append(college_new.columns[0])
    college_new["State"] = state_cont
    college_new = college_new.set_index(["State",college_new.index])
    college_new.columns=["College"]
    college_data = college_data.append(college_new)
college_data   
```

```python
#Locate the first state in our data
hs_data = pd.DataFrame(hs_df.T.loc["Alaska",:])

#container for loop appending
hs_cont = []

#Loop to get repeating state name in column and setting up new loop
for x in range(len(hs_data)):
    hs_cont.append(hs_data.columns[0])
hs_data["State"] = hs_cont
hs_data = hs_data.set_index(["State",hs_data.index]) 
hs_data.columns=["High School"]

#Iterate through all states to get data in a preferred format for mongodb
for state in state_list:

    hs_new = pd.DataFrame(hs_df.T.loc[state,:])

    hs_cont = []

    for x in range(len(hs_new)):
        hs_cont.append(hs_new.columns[0])
    hs_new["State"] = hs_cont
    hs_new = hs_new.set_index(["State",hs_new.index])
    hs_new.columns=["High School"]
    hs_data = hs_data.append(hs_new)
hs_data  
```

```python
#Save finished file into a csv
combo_df = pd.merge(college_data, hs_data, how = "outer", on = ["State","DATE"])
level=1
combo_df.index = combo_df.index.set_levels(combo_df.index.levels[level].astype(str),level=level)
combo_df.to_csv("../data/education.csv")
```
